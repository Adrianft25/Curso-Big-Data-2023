```python
#Rating,Company,Average,Lowest,Highest,yr/mo/hr

lineaOrigen = ["3.7","Optum","1508702","8,87,265","24,00,000","/yr"]
#lineaOrigen = ["4.1","Electronic Arts","775143","9,23,568","10,14,289","/yr"]
#lineaOrigen = ["","Hritik","727941","8,75,612","9,44,241","/yr"]

tipo = 0
high = 0
low = 0
media = 0
enEuros = 0

lineaOrigen[1] = lineaOrigen[1].split(" ")

if len(lineaOrigen[1]) == 1:

  nomC = lineaOrigen[1][0]

  if float(lineaOrigen[0]) < 3.5:
    tipo = 1
  elif float(lineaOrigen[0]) >= 3.5 and float(lineaOrigen[0]) < 4:
    tipo = 2
  elif float(lineaOrigen[0]) >= 4 and float(lineaOrigen[0]) < 4.5:
    tipo = 3
  elif float(lineaOrigen[0]) >= 4.5:
    tipo = 4
  else:
    tipo = "ERROR"

  lineaOrigen[-3] = lineaOrigen[-3].split(",")

  if len(lineaOrigen[-3][1]) == 2:
    lineaOrigen[-3] = ",".join(lineaOrigen[-3][:1] + ["0"] + lineaOrigen[-3][1:])

    low = float(lineaOrigen[-3].replace(",", ""))

  elif len(lineaOrigen[-3][1]) == 1:
    lineaOrigen[-3] = ",".join(lineaOrigen[-3][:1] + ["0"] + ["0"] + lineaOrigen[-3][1:])

    low = float(lineaOrigen[-3].replace(",", ""))

  else:
    low = "ERROR"
  
  lineaOrigen[-2] = lineaOrigen[-2].split(",")

  if len(lineaOrigen[-2][1]) == 2:

    lineaOrigen[-2] = ",".join(lineaOrigen[-2][:1] + ["0"] + lineaOrigen[-2][1:])

    high = float(lineaOrigen[-2].replace(",", ""))

  elif len(lineaOrigen[-2][1]) == 1:

    lineaOrigen[-2] = ",".join(lineaOrigen[-2][:1] + ["0"] + ["0"] + lineaOrigen[-2][1:])

    high = float(lineaOrigen[-2].replace(",", ""))

  else:
    high = "ERROR"
  
  if high == "ERROR" or low == "ERROR":
    media = "ERROR"
  else:
    media = round((low + high) / 2 ,2)

  if media != "ERROR":
    if lineaOrigen[-1] == "/mo":
      media = media * 12
    elif lineaOrigen[-1] == "/hr":
      media = media * 218 * 8
    enEuros = round(media * 0.011,2)

  else:
    enEuros = "ERROR"

else:
  nomC = "ERROR"


lineaDestino = [nomC, tipo, media, enEuros]
#lineaDestino = [nomC, tipo, media, enEuros, lineaOrigen[-1]]


print(lineaDestino)
```
