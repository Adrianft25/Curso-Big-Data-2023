1. Calcular la hipotenusa de un triangulo
2. Fórmula para la hipotenusa: h = sqrt(a2 + b2)

```python
'''
Autor: Adrián Freire Torres
10/04/2023
adrianft25@gmail.com
Cálculo de la hipotenusa de un triangulo
'''

import math

print("CÁLCULO DE LA HIPOTENUSA DE UN TRIÁNGULO")

#in
a = float(input("Introduce la medida del primer cateto: "))
b = float(input("Introduce la medida del segundo cateto: "))

#out
h = math.sqrt(a ** 2 + b ** 2)

print("El valor de la hipotenusa es: ", round(h, 2))
```

Haz una calculadora básica pida al usuario dos valores, a y b.

Según la opción que desean, realizar la operación:

- Si operación es 1 entonces debemos ver el resultado de a + b
- Si operación es 2 entonces debemos ver el resultado de a * b
- Si operación es 3 entonces debemos ver el resultado de a - b
- Si operación es 4 entonces debemos ver el resultado de a / b

```python
'''
Autor: Adrián Freire Torres
10/04/2023
adrianft25@gmail.com
Estructuras de Control | Calculadora

=============================================

opcion = operación a realizar
a = primer numero
b = segundo numero
total = resultado de la operación
'''

print("CALCULADORA")

opcion = int(input("Selecciona tu operación:" + "\n" + "1. Suma"+ "\n" + "2. Multiplicación"+ "\n" + "3. Resta "+ "\n" + "4. División" + "\n"))
a = float(input("Introduce el primer numero: "))
b = float(input("Introduce el segundo numero: "))

if(opcion == 1):
  total = a + b

elif(opcion == 2):
  total = a * b

elif(opcion == 3):
  total = a - b

elif(opcion == 4):
  while b == 0:
    b = float(input("No se puede dividir entre cero, introduce otro número: "))
  total = a / b

print("Total:", round(total, 2))
```

Si es mayor a 100 y menor a 150 aplicamos un 12%. Si tienes 150 un 15% y sino, el resto de los casos de más de 150, un 20%. ¡Crea la variable puntos y juega con ella!

```python
'''
Autor: Adrián Freire Torres
10/04/2023
adrianft25@gmail.com
Estructuras de Control | Calculadora de Descuentos por Puntos

=============================================

puntos = cantidad de puntos para realizar el descuento
precioCompra = precio de la compra al que se aplicará el descuento
descuento = descuento que se hará
precioFinal = precio con el descuento hecho

'''

print("CALCULADORA DE DESCUENTOS POR PUNTOS")

precioCompra = float(input("Introduce el precio final de compra: "))
puntos = int(input("Introduce tus puntos: "))

if puntos >= 0 and puntos <= 100:
  descuento = precioCompra * 0.10
elif puntos > 100 and puntos < 150:
  descuento = precioCompra * 0.12
elif puntos == 150:
  descuento = precioCompra * 0.15
elif puntos > 150:
  descuento = precioCompra * 0.20
else:
  print("ERROR. INTRODUCE UN NUMERO VÁLIDO")

precioFinal = precioCompra - descuento

print("El precio final de compra es de :", round(precioFinal, 2), "€")
```

Desarrollar un programa en Python que determine si el año introducido es un año bisiesto o no. Sabiendo que…

1. Si el año es uniformemente divisible por 4, vaya al paso 2. De lo contrario, vaya al paso 5.
2. Si el año es uniformemente divisible por 100, vaya al paso 3. De lo contrario, vaya al paso 4.
3. Si el año es uniformemente divisible por 400, vaya al paso 4. De lo contrario, vaya al paso 5.
4. El año es un año bisiesto (tiene 366 días).
5. El año no es un año bisiesto (tiene 365 días).

```python
'''
Autor: Adrián Freire Torres
10/04/2023
adrianft25@gmail.com
Estructuras de Control | Años bisiestos

=============================================

anho = año introducido por el usuario

'''

print("AÑOS BISIESTOS")

anho = int(input("Introduce un año: "))

if anho % 4 == 0:
  if anho % 100 == 0:
    if anho % 400 == 0:
      print("El año:", anho, "es bisiesto")
    else:
      print("El año:", anho, "NO es bisiesto")
  else:
    print("El año:", anho, "es bisiesto")
else:
  print("El año:", anho, "NO es bisiesto")
```

Partiendo de la tarifa anual (que puede cambiar), nos piden que debemos calcular el precio de la tarifa de nuestro polideportivo, sabiendo las siguientes condiciones:
