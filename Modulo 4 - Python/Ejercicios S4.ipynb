```python
'''
Autor: Adrián Freire Torres
13/04/2023
adrianft25@gmail.com
Estructuras de Control | Calculo de hipotenusa

=============================================

cateto1 = lado 1 del triangulo
cateto2 = lado 2 del triangulo
hipotenusa = hipotenusa del triangulo
'''

import math

print("CÁLCULO DE LA HIPOTENUSA")

cateto1 = float(input("Introduce la medida del primer cateto: "))
cateto2 = float(input("Introduce la medida del segundo cateto: "))


while cateto1 <= 0 or cateto2 <= 0:
  print("ERROR. NINGUNO DE LOS CATETOS PUEDE SER MENOR O IGUAL A CERO")
  cateto1 = float(input("Introduce la medida del primer cateto: "))
  cateto2 = float(input("Introduce la medida del segundo cateto: "))

hipotenusa = math.sqrt(cateto1 ** 2 + cateto2 ** 2)
print("La hipotenusa vale:", round(hipotenusa,2))
  
```
